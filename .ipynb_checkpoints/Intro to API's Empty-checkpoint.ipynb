```python
url = 'https://ergast.com/api/f1/2008/5/driverStandings.json'
```

```python
#response object
response = requests.get(url) # There are four main methods of how we interact with data, most of the time we use 'get' basically saying I want to get this information
#The requests library is going to go to the url, pull it back and save all of the information that it gets in response

print(response) # When you print response you get back 200, view the status code 200 here https://developer.mozilla.org/en-US/docs/Web/HTTP/Status
# 200 series is basically a successful response
# If you didnt make the API and you get a 500 back then its not your fault, however If you made the API and you get a 500 back then its your fault


```

```python
#check and make sure we got a successful response from the API

response.ok # This is a way to check and make sure we got a response back 
```

```python
# view the response as JSON

# Rather than seeing <Response [200]> we want to see JSON
response.json() #This is a function which is why there are parenthesis
# This is the data for the year 2008 standings
# JSON Vue is how your able to view the code in more clearly
```

```python
# What data type are we getting back
type(response.json())
#it returns a dictionary of course because thats how pythons going to read that
# if you wanted to include an image of the person in the data because this is text only (above) you can add the image to another website and give it a unique url
```

```python
# We only want the Driver Standings 
#Now were going to parsing through the data 
# data = response.json()['MRData'] # This is the first dictionary (there is a dictionary within a dictionary)
# # data
# # However were trying to target the DriverStandings key which has a value of another dictionary
# data2 = response.json()['MRData']['StandingsTable']
# # data2

# data3 = response.json()['MRData']['StandingsTable']['StandingsLists']
# data3

for x in range(5): # without the for loop 
    data4 = response.json()['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings'][x]['Driver']['familyName'] #This will give you the data for driver one and below
    print(data4) 


my_racer_data = response.json()['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']
my_racer_data
```

```python
def get_racer_info(data):
    new_data = []
    for racer in data:
        racer_dict = {}
        racer_name = f'{racer["Driver"]["givenName"]} {racer["Driver"]["familyName"]}'
        
        racer_dict[racer_name]={ # All of the data is not needed in the 'data' listed above so we can grab the specific keys and values just like the example below
            'first_name': racer ['Driver']['givenName'],
            'last_name': racer ['Driver']['familyName'],
            'position': racer['position'],
            'wins': racer['wins'],
            'DOB': racer['Driver']['dateOfBirth'],
            'nationality': racer['Driver']['nationality'],
            'constructor': racer['Constructors'][0]['name']                      
        }
        new_data.append(racer_dict)
    return new_data
                                
                                
get_racer_info(my_racer_data)

```

```python
def get_driver_info_by_year_and_round(year, r):
    url = f"https://ergast.com/api/f1/{year}/{r}/driverStandings.json" # everytime we run the function, Whichever year and round is passed in will be given
    # to the url and it will go find the page for us. You dont have to load the page in a browser
    response = requests.get(url)
    if not response.json()['MRData']['StandingsTable']['StandingsLists']: # If theres data in here, this will be true, if not
        return "We had an error loading your data 🙃😂🤮 It is likely the year or round is not in the database " #ctrl, cmmd, space emoji mac shortcut
    data = response.json()['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']
    return get_racer_info(data)

get_driver_info_by_year_and_round(2019, 4)
    
```
